```python
# 중간에 잘린 부분이 있어서 확실하진 않음.
VIDEO_NAME = "rock-paper-scissors"
FRAME_RANGE = range(146232, 148486, 50)
# 댓글 창이 화면 왼쪽에서 몇 픽셀 떨어져 있는지에 대한 자료
COMMENTS_FRAME_LOCATION = int(563 * 0.75)
ORIGINAL_RESULTS = "김풍"
```

```python
# 댓글에서 무엇을 가져오는지 결정. rock-paper-scissors를 제외하면 모두 1과 2.
if VIDEO_NAME == "rock-paper-scissors":
    COUNTING_VALUES = ("1", "2", "3")
else:
    COUNTING_VALUES = ("1", "2")
```

```python
# 배열 유사도 계산. 후에 같은 이미지를 스킵할 때 사용.

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


def get_multidimensional_similarity(array1, array2, threshold: float | None = None):
    # sourcery skip
    """코사인 유사도를 계산하지만 완전히 같은 배열을 입력해도 1.0이 아닐 수 있고 1.0을 넘길 수도 있음."""
    # 배열을 형상 (-1, 1)인 2차원 배열로 변환
    reshaped_array1 = array1.reshape(-1, 1)
    reshaped_array2 = array2.reshape(-1, 1)

    # 코사인 유사도 계산
    cos_sim = cosine_similarity(reshaped_array1.T, reshaped_array2.T)[0, 0]
    if threshold is None:
        return cos_sim
    else:
        return cos_sim >= threshold
```

```python
import os
import shutil
from contextlib import suppress

chatting_images_path = f"chatting-images/{VIDEO_NAME}"
# chatting_ocr_results_path = f"chatting-ocr-results/{VIDEO_NAME}"

with suppress(Exception):
    shutil.rmtree(chatting_images_path)
    # shutil.rmtree(chatting_ocr_results_path)

os.mkdir(chatting_images_path)
# os.mkdir(chatting_ocr_results_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from typing import *

cap = cv2.VideoCapture(f"videos/{VIDEO_NAME}.mp4")
prev_commentsframe = np.zeros((1080, COMMENTS_FRAME_LOCATION, 3))

for i in FRAME_RANGE:
    cap.set(cv2.CAP_PROP_POS_FRAMES, i)
    ret, frame = cap.read()
    frame = np.array(frame)
    if not ret:
        break

    commentsframe = frame[:, -COMMENTS_FRAME_LOCATION:, :]

    if get_multidimensional_similarity(prev_commentsframe, commentsframe, 0.95):
        print(f"frame no. {i} is not captured; it's same with previous frame.")
    else:
        cv2.imwrite(f"{chatting_images_path}/{i:04d}.png", commentsframe)

    prev_commentsframe = commentsframe

cap.release()
```

```python
import easyocr

reader = easyocr.Reader(["ko", "en"])
```

```python
import os

results = []
directory = f"chatting-images/{VIDEO_NAME}/"
for filename in os.listdir(directory):
    print(filename)
    results.append(reader.readtext(directory + filename, detail=0))

results
```

```python
def vote_result_of_one_comment(comment: str, counts=("1", "2")):
    count_results = {count: comment.count(count) for count in counts}

    max_val = max(count_results.values())
    if list(count_results.values()).count(max_val) >= 2:
        return None

    return max(count_results.items(), key=lambda x: x[1])[0]
```

```python
from collections import defaultdict

vote_results = defaultdict(int)
for comments in results:
    prev_comment_result = None
    for comment in comments:
        comment_result = vote_result_of_one_comment(
            comment, counts=COUNTING_VALUES)
        # print(comment, comment_result)
        if prev_comment_result != comment_result and comment_result is not None:
            vote_results[comment_result] += 1
        prev_comment_result = comment_result

vote_results
```

```python
import json

with open(f"chatting-ocr-results/{VIDEO_NAME}.json", "w", encoding="utf-8") as f:
    json.dump((results, vote_results), f, ensure_ascii=False)

with open("results.txt", "a", encoding="utf-8") as f:
    total = sum(vote_results.values())
    if COUNTING_VALUES == ("1", "2"):
        chim = vote_results["1"]
        pearl = vote_results["2"]
        if chim == pearl:
            winner_announcement = "동률 무승부입니다."
        elif chim > pearl:
            winner_announcement = f"침착맨이 {chim - pearl}표({chim / total * 100 - pearl / total * 100:.0f}%p) 앞서 승리입니다."
        else:
            winner_announcement = f"주호민이 {pearl - chim}표({pearl / total * 100 - chim / total * 100:.0f}%p) 앞서 승리입니다."
        print(
            f"{VIDEO_NAME}의 투표 결과는 침착맨 {chim}표({chim / total * 100:.0f}%) 대 주호민 {pearl}({pearl / total * 100:.0f}%)표로 "
            + winner_announcement
            + f' 이는 기존에 {ORIGINAL_RESULTS}을 우승으로 집계하였던 것과 {"상이합니다." if chim < pearl and ORIGINAL_RESULTS == "침착맨" else "일치합니다."}',
            file=f,
        )
    elif COUNTING_VALUES == ("1", "2", "3"):
        chim = vote_results["1"]
        pearl = vote_results["2"]
        pung = vote_results["3"]
        if len({chim, pearl, pung}) in {1, 2}:
            winner_announcement = f"동률 무승부입니다. ({4 - len({chim, pearl, pung})}명 동률)"
        else:
            best, runner_up, _ = sorted(((chim, "침착맨"), (pearl, "주호민"), (pung, "김풍")), reverse=True)
            diff = best[0] - runner_up[0]
            diff_percent_point = diff / total * 100
            # 공교롭게도 모두 받침이 있어서 '이' 사용 가능
            winner_announcement = f"{best[1]}이 {diff}표({diff_percent_point:.0f}%p) 앞서 승리입니다."
            print(
                f"{VIDEO_NAME}의 투표 결과는 침착맨 {chim}표({chim / total * 100:.0f}%) 대 주호민 {pearl}({pearl / total * 100:.0f}%)표 " f"대 김풍 {pung}({pung / total * 100:.0f}%)표로 " + winner_announcement
                # 이는 그냥 김풍을 우승으로 집계함.
                + f' 이는 김풍을 우승으로 집계하였던 것과 {"상이합니다." if best[1] != "김풍" else "일치합니다."}',
                file=f,
            )
    else:
        print("Unknown COUNTING_VALUES, check `chatting-ocr-images` instead. " f"{VIDEO_NAME = }, {COUNTING_VALUES = }, {vote_results = }", file=f)
```
